```python
 # Data Cleaning
```

```python
# To fill in the missing values in gender, use mode
gender_mode= df["Gender"].mode()
```

```python
gender_mode
```

```python
df.Gender.fillna(gender_mode, inplace=True)
```

```python
df
```

```python
df.query("Married=='Yes'")["Loan_Status"].value_counts
```

```python
# To fill Married, use mode
married_mode= df['Married'].mode()
df.Married.fillna(married_mode, inplace= True)
```

```python
# To clean and fill Dependents,use mode
dependents_mode= df["Dependents"].mode()
df.Dependents.fillna(dependents_mode, inplace= True)
```

```python
# For missing values in Self_ employed, use mode
self_employed_mode = df["Self_Employed"].mode()
df.Self_Employed.fillna(self_employed_mode, inplace= True)
```

```python
# For LoanAmount, use mean
loanamount_mean = df["LoanAmount"].mean
df.LoanAmount.fillna(loanamount_mean, inplace = True)
```

```python
# Loan_Amount_Term, use mean
loan_amount_term_mean = df["Loan_Amount_Term"].mean
df.Loan_Amount_Term.fillna(loan_amount_term_mean, inplace= True)
```

```python
# For Credit History, use mean
credit_history_mean = df["Credit_History"].mean
df.Credit_History.fillna(credit_history_mean, inplace = True)
```
